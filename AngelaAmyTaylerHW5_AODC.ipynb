b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'], which you may have to convert date type again, put those years as labels, and set the titles for each small plot to their state names. In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years.

```python
#your code goes here

import seaborn as sns
import matplotlib.pyplot as plt

house['date'] = pd.to_datetime(house['date'])

# Select rows for the specified states
selected_states = ['Michigan', 'California', 'Texas', 'Indiana']
selected_rows = house[house['state'].isin(selected_states)].copy()

# Convert 'date' to datetime using .loc to modify the original DataFrame
selected_rows.loc[:, 'date'] = pd.to_datetime(selected_rows['date'])

# Preview the new dataframe
selected_rows.head()



# Set the style
sns.set(style="whitegrid")

# Create a FacetGrid
g = sns.FacetGrid(selected_rows, col="state", col_wrap=2, height=5, palette="tab20")

# Map the line plots
g.map(sns.lineplot, "date", "house_price_index", color="red")

# Set x-axis ticks and labels
g.set(xticks=pd.to_datetime(['1980-01-01', '2000-01-01', '2020-01-01']), xticklabels=['1980', '2000', '2020'])

# Set y-axis label
g.set_ylabels("House Price Index")

# Set titles for each plot
g.set_titles(col_template="{col_name}")

# Remove x-axis label
g.set_xlabels("")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
